```python
Sparta = pd.read_csv("Sparta0.csv", sep=";", encoding="utf-8")
Sparta.info ()
```

```python
print(Sparta["Date"].dtype)
Sparta["Date"] = pd.to_datetime(Sparta["Date"], dayfirst=True)
print(Sparta["Time"].dtype)
print(Sparta["Time"].unique())
```

```python

Time_text_clean = Sparta["Time"].str.replace("odp.", "PM")
Time_clean = Time_text_clean.str.strip()
```

```python
Time_final = pd.to_datetime(
    Time_clean, 
    format="%I:%M %p", 
    errors='coerce'
).dt.time

Sparta["Time"] = Time_final
Sparta
```

```python
print(Sparta["Home team"].dtype)
Sparta["Home team"] = Sparta["Home team"].astype(str).str.split("(").str[0].str.strip()
Sparta["Away team"]= Sparta["Away team"].astype(str).str.split("(").str[0].str.strip()
Sparta["Home team"] = Sparta["Home team"].replace(r'^[0-9.\s]+', '', regex=True)
Sparta["Away team"] = Sparta["Away team"].replace(r'^[0-9.\s]+', '', regex=True)
Sparta
```

```python
sezony = []

for datum in Sparta['Date']:
    rok = datum.year
    mesic = datum.month
    
    if mesic >= 7:
        sezona = str(rok) + "/" + str(rok + 1)
    else:
        sezona = str(rok - 1) + "/" + str(rok)
    
    sezony.append(sezona)

Sparta['Season'] = sezony
Sparta
```

```python
print(Sparta["Attendance"].dtype)
Sparta["Attendance"] = Sparta["Attendance"].astype(str).str.replace(".", "", regex=False)
nova_cisla = []

for hodnota in Sparta["Attendance"]:
    text_hodnoty = str(hodnota)
    if text_hodnoty == "x" or text_hodnoty == "nan":
        nova_cisla.append(0)
    else:
        nova_cisla.append(int(text_hodnoty))
Sparta["Attendance"] = nova_cisla
Sparta
```

```python
prazdne_casy = Sparta[Sparta["Time"].isna()]
prazdne_casy
```

```python
Sparta["Time"] = Sparta["Time"].fillna("15:00:00")
```

```python
Sparta = Sparta.drop(["Result", "System of play"], axis=1)
```

```python
nulove_zápasy = Sparta[(Sparta["Attendance"] == 0) | (Sparta["Attendance"].isna())]
nulove_zápasy
```

```python
import pandas as pd
import numpy as np

Sparta_pocas = pd.read_csv("MODEL_FINAL.csv")


Sparta_pocas['temp_comfort'] = np.where(Sparta_pocas['temperature_2m (°C)'] < 0, 'Mrazy',
                               np.where(Sparta_pocas['temperature_2m (°C)'] < 10, 'Zima',
                               np.where(Sparta_pocas['temperature_2m (°C)'] < 20, 'Chladno',
                               np.where(Sparta_pocas['temperature_2m (°C)'] < 28, 'Ideálně', 'Horko'))))


Sparta_pocas['is_extreme_weather'] = np.where((Sparta_pocas['precipitation (mm)'] > 3) | (Sparta_pocas['snowfall (cm)'] > 0), 1, 0)

# srážek
Sparta_pocas['weather_type'] = np.where(Sparta_pocas['snowfall (cm)'] > 0, 'Sněžení',
                               np.where(Sparta_pocas['rain (mm)'] > 2, 'Silný déšť',
                               np.where(Sparta_pocas['rain (mm)'] > 0, 'Mírný déšť', 'Beze srážek')))

#  nepohodli
Sparta_pocas['weather_discom'] = (
    (Sparta_pocas['temperature_2m (°C)'] < 5).astype(int) + 
    (Sparta_pocas['rain (mm)'] > 1).astype(int) + 
    (Sparta_pocas['snowfall (cm)'] > 0).astype(int)
)


Sparta_pocas = pd.get_dummies(Sparta_pocas, columns=['temp_comfort', 'weather_type'])

Sparta_pocas.to_csv("MODEL_FINAL.csv", index=False, encoding="utf-8-sig")
```

```python
import numpy as np
import pandas as  pd
df= pd.read_csv("SPARTA_FINAL.csv")

def temp_category(temp):
    if pd.isna(temp):
        return np.nan
    if 15 <= temp <= 24:
        return 1
    elif 10 <= temp < 15 or 24 < temp <= 27:
        return 2
    elif 5 <= temp < 10 or 27 < temp <= 30:
        return 3
    elif 0 <= temp < 5 or 30 < temp <= 32:
        return 4
    else:
        return 5


def precip_category(precipitation):
    if pd.isna(precipitation):
        return np.nan
    if precipitation == 0:
        return 1
    elif 0 < precipitation <= 0.2:
        return 2
    elif 0.2 < precipitation <= 1:
        return 3
    elif 1 < precipitation <= 3:
        return 4
    else:
        return 5


df["TEMP_KATEGORIE"] = df["TEMPERATURE"].apply(temp_category)
df["SRAZKY_KATEGORIE"] = df["PRECIPITATION"].apply(precip_category)

df["POCASI_KATEGORIE"] = df[["TEMP_KATEGORIE", "SRAZKY_KATEGORIE"]].max(axis=1)

df["POCASI_LABEL"] = df["POCASI_KATEGORIE"].map({
    1: "1_Ideální počasí",
    2: "2_V pohodě",
    3: "3_Hraniční komfort",
    4: "4_Nepříjemné",
    5: "5_Extrém"
})



df = df.drop(columns=['IS_EXTREME_WEATHER','WEATHER_DISCOM','TEMP_COMFORT_CHLADNO','TEMP_COMFORT_HORKO','TEMP_COMFORT_IDO','TEMP_COMFORT_MRAZY','TEMP_COMFORT_ZIMA','WEATHER_TYPE_NORAIN','WEATHER_TYPE_SLIGHT','WEATHER_TYPE_SNOW'], errors='ignore')

df.to_csv("SPARTA_FINAL1.csv", index=False, encoding="utf-8-sig")
```
